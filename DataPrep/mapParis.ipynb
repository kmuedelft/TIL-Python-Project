```python
import folium

# Define the latitude and longitude range
lat_min, lat_max = 48.65, 49.1
lon_min, lon_max = 2.01, 2.76

# Create a base map centered around the midpoint of the bounding box
center_lat, center_lon = (lat_min + lat_max) / 2, (lon_min + lon_max) / 2
m = folium.Map(location=[center_lat, center_lon], zoom_start=10)

# Define the bounding box corners
bounding_box = [
    [lat_min, lon_min],  # Bottom-left corner
    [lat_min, lon_max],  # Bottom-right corner
    [lat_max, lon_max],  # Top-right corner
    [lat_max, lon_min],  # Top-left corner
    [lat_min, lon_min]   # Closing the box (back to bottom-left)
]

# Add the bounding box to the map
folium.PolyLine(bounding_box, color="red", weight=2.5).add_to(m)

# Display the map
m
```
