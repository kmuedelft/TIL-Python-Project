```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df319 = pd.read_pickle('A319_final.pkl')
df320 = pd.read_pickle('A320_final.pkl')    
df321 = pd.read_pickle('A321_final.pkl')
df332 = pd.read_pickle('A332_final.pkl')
```

```python
# Extract the correlation values for climb_rate from each dataframe
corr_df319 = df319.corr().loc['climb_rate']
corr_df320 = df320.corr().loc['climb_rate']
corr_df321 = df321.corr().loc['climb_rate']
corr_df332 = df332.corr().loc['climb_rate']

# Create a new DataFrame with one row for each airplane
climb_rate_correlation_all = pd.DataFrame({
    'A319': corr_df319,
    'A320': corr_df320,
    'A321': corr_df321,
    'A332': corr_df332
}).T

climb_rate_correlation_all

# make a heatmap of the correlation values
plt.figure(figsize=(16, 2))
sns.heatmap(climb_rate_correlation_all, annot=True, cmap='coolwarm')
plt.title('Correlation between climb rate and other parameters')
plt.show()

```
